## Set A Solution

```python
import numpy as np

def sol_quiz1_setA(M, t):
  cur_row = 0
  cur_col = 0
  n, m = M.shape

  dx = [0, 1, 0, -1]
  dy = [1, 0, -1, 0]

  score = M[0][0]

  while t > 0:
    mx = -1
    next_row = -1
    next_col = -1
    for i in range(len(dx)):
      if 0 <= cur_row + dx[i] < n and 0 <= cur_col + dy[i] < m:
        if M[cur_row + dx[i]][cur_col + dy[i]] > mx:
          mx = M[cur_row + dx[i]][cur_col + dy[i]]
          next_row = cur_row + dx[i]
          next_col = cur_col + dy[i]
    cur_row = next_row
    cur_col = next_col
    score += M[cur_row][cur_col]
    t -= 1

  return score
```

```python
sol_quiz1_setA(np.array([[1, 2, 3], [4, 5, 6], [9, 3, 7]]), 4)
```

## Set B Solution

```python
import numpy as np

def sol_quiz1_setB(M, t):
  cur_row = 0
  cur_col = 0
  n, m = M.shape

  dx = [0, 1, 0, -1]
  dy = [1, 0, -1, 0]

  score = M[0][0]

  while t > 0:
    mn = 100
    next_row = -1
    next_col = -1
    for i in range(len(dx)):
      if 0 <= cur_row + dx[i] < n and 0 <= cur_col + dy[i] < m:
        if M[cur_row + dx[i]][cur_col + dy[i]] < mn:
          mn = M[cur_row + dx[i]][cur_col + dy[i]]
          next_row = cur_row + dx[i]
          next_col = cur_col + dy[i]
    cur_row = next_row
    cur_col = next_col
    score += M[cur_row][cur_col]
    t -= 1

  return score
```
