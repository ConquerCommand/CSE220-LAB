```python
class Node:
  def __init__(self,elem,next):
    self.elem=elem
    self.next=next
```

```python
#Set-A
def printList(head):
  temp=head
  while(temp!=None):
    print(temp.elem,end=" ")
    temp=temp.next
  print()

def removeDuplicates(head):
  temp=head
  count=0
  while(temp!=None):
    if(temp.next==None):
      break
    if (temp.elem==temp.next.elem):
      temp.next=temp.next.next
      count+=1
    temp=temp.next

  print("Removed Items: ",count)
  print("Modified List: ",end="")
  printList(head)



# DRIVER CODE
n1=Node(10,None)
n2=Node(20,None)
n3=Node(30,None)
n4=Node(10,None)
n5=Node(10,None)
n6=Node(50,None)
n7=Node(50,None)
n1.next=n2
n2.next=n3
n3.next=n4
n4.next=n5
n5.next=n6
n6.next=n7

removeDuplicates(n1)

# MARKING RUBRIC
# Loop = 1 marks
# Loop Conditions = 3 marks
# Duplicate Check Conditions = 3 marks
# Updating Node Link to Remove = 2 marks
# Count for Duplicate Nodes = 2 marks
# Updating Temp For Loop = 1 marks
# printList = 3 marks
```
