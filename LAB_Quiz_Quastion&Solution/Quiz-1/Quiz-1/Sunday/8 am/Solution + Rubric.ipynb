```python
# SET A
import numpy as np
def row_max(arr, row):
    max = arr[row][0]
    for i in range(1, len(arr[0])):
        if max < arr[row][i]:
            max = arr[row][i]
    return max

def col_max(arr, col):
    max = arr[0][col]
    for i in range(1, len(arr)):
        if max < arr[i][col]:
            max = arr[i][col]
    return max

def replace_array(arr):
    for r in range(len(arr)):
        r_max = row_max(arr, r)
        for c in range(len(arr[0])):
            c_max = col_max(arr, c)
            if r_max < c_max:
                arr[r][c] = r_max
            else:
                arr[r][c] = c_max
    return arr

arr = np.array([[7, 5, 6],
                [1, 8, 2],
                [9, 8, 7],
                [2, 8, 1],
                [4, 7, 3]], dtype = int)

print(replace_array(arr))

```
