```python
#Run this cell
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next

def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

```python
def count(head):
    length = 0
    current = head
    while current:
        length += 1
        current = current.next
    return length
```

```python
def nodeAt(head, index):
    current = head
    for _ in range(index):
        if current is None:
            return None
        current = current.next
    return current
```

```python
def sort(head):
  temp = head
  while temp.next!=None:
    if temp.elem>temp.next.elem:
      return False
    temp=temp.next
  return True
```

```python
def is_sorted(head):
    n = count(head)  # Get the length of the linked list
    mid = n // 2
    list1 = head
    list2 = nodeAt(head,mid+1)
    cnt = 0
    while cnt<mid:
      if cnt == mid-1:
          return "First"
      elif list1.elem> list1.next.elem:
          break
      list1 = list1.next
      cnt+=1
    while True:
      x = sort(list2)
      if x:
        return "Last"
      else:
        return None
```

```python
#Driver Code
head1 = createList([1,2,5,9,3])
head2 = createList([8,2,7,9,5,6,7])
head3 = createList([10,1,5,6,2])
print(is_sorted(head3))
```

```python
def sum_check(head):
  sum1 = 0
  sum2 = 0
  n = count(head)
  mid = n // 2
  list1 = head
  list2 = nodeAt(head,mid+1)
  cnt = 0
  while cnt<mid:
    sum1+=list1.elem
    list1 = list1.next
    cnt+=1

  while list2:
    sum2+=list2.elem
    list2 = list2.next
  if sum1==sum2:
    return True
  else:
    return False
```

```python
#Driver Code
head1 = createList([1,2,5,2,1])
head2 = createList([8,2,7,9,5,6,7])
print(sum_check(head2))
```
