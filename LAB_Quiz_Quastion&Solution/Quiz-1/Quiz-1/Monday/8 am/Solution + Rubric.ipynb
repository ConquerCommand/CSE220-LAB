```python
import numpy as np
```

```python
#Set A
def replace_array_A(arr):
  row, col = arr.shape
  row_sum = np.zeros(row, dtype=int)
  col_sum = np.zeros(col, dtype=int)
  for i in range(row):
    for j in range(col):
      row_sum[i] += arr[i][j]
      col_sum[j] += arr[i][j]
  for i in range(row):
    for j in range(col):
      arr[i][j] = (row_sum[i] - arr[i][j]) * (col_sum[j] - arr[i][j])
  return arr

arr = np.array([[1, 2, 3],
                [4, 5, 6],
                [9, 8, 7],
                [2, 8, 1]], dtype = int)
output = replace_array_A(arr)
print(output)
```

```python
def replace_array_B(arr):
  row, col = arr.shape
  row_prod = np.ones(row, dtype=int)
  col_prod = np.ones(col, dtype=int)
  for i in range(row):
    for j in range(col):
      row_prod[i] *= arr[i][j]
      col_prod[j] *= arr[i][j]
  for i in range(row):
    for j in range(col):
      arr[i][j] = (row_prod[i] / arr[i][j]) + (col_prod[j] / arr[i][j])
  return arr

arr = np.array([[1, 2, 3],
                [4, 5, 6],
                [9, 8, 7],
                [2, 8, 1]], dtype = int)
output = replace_array_B(arr)
print(output)
```
